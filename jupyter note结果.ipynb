```python
txt=txt.lower()
```

```python
for ch in';“”‘’.,!?():#*[]\|/-_':
    txt=txt.replace(ch,"  ")
```

```python
words=txt.split()
```

```python
 len(set(words))
```

```python
counts={}
```

```python
for word in words :
      counts[word]= counts.get(word,0)+1 
```

```python
counts
```

```python
items= list(counts.items())
```

```python
items
```

```python
items.sort(key=lambda x:x[1],reverse=True)
```

```python
items
```

```python
for i in range(100):
       print(items[i][0], items[i][1])
```
